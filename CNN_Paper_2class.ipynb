```python
import pickle
file_to_read = open("/content/drive/MyDrive/NLP/X_train_4", "rb")
X = pickle.load(file_to_read)
file_to_read = open("/content/drive/MyDrive/NLP/y_train_4", "rb")
y = pickle.load(file_to_read)
```

```python
X_train = X[:77000]
X_test = X[77000:]
y_train = y[:77000]
y_test = y[77000:]
```

```python
import tensorflow as tf
import keras 
title_input = keras.Input(
    shape=(50,100), name="title"
)
y1=keras.layers.Conv1D(32,2,activation='relu',input_shape=(50,100))(title_input)
y2=keras.layers.Conv1D(32,2,activation='relu',input_shape=(50,100))(title_input)
y3=keras.layers.Conv1D(32,3,activation='relu',input_shape=(50,100))(title_input)
y4=keras.layers.Conv1D(32,3,activation='relu',input_shape=(50,100))(title_input)
y5=keras.layers.Conv1D(32,4,activation='relu',input_shape=(50,100))(title_input)
y6=keras.layers.Conv1D(32,4,activation='relu',input_shape=(50,100))(title_input)

m1=tf.keras.layers.MaxPooling1D(pool_size=49,strides=1)(y1)
m2=tf.keras.layers.MaxPooling1D(pool_size=49,strides=1)(y2)
m3=tf.keras.layers.MaxPooling1D(pool_size=48,strides=1)(y3)
m4=tf.keras.layers.MaxPooling1D(pool_size=48,strides=1)(y4)
m5=tf.keras.layers.MaxPooling1D(pool_size=47,strides=1)(y5)
m6=tf.keras.layers.MaxPooling1D(pool_size=47,strides=1)(y6)

concanate = tf.keras.layers.concatenate([m1,m2,m3,m4,m5,m6])

den_lay1 = tf.keras.layers.Dense(256, name="MLP1")(concanate)

den_lay2 = tf.keras.layers.Dense(128, name="MLP2")(den_lay1)

den_lay3 = tf.keras.layers.Dense(64, name="MLP3")(den_lay2)

den_lay4 = tf.keras.layers.Dense(32, name="MLP4")(den_lay3)

den_lay5 = tf.keras.layers.Dense(16, name="MLP5")(den_lay4)

op_lay = tf.keras.layers.Dense(2, activation='softmax',name="priority")(den_lay5)

model = keras.Model(
    inputs=[title_input],
    outputs=[op_lay],
)

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.fit(X_train,y_train,epochs=10)
```

```python
score=model.evaluate(X_test,y_test)
print(score[1])
```
